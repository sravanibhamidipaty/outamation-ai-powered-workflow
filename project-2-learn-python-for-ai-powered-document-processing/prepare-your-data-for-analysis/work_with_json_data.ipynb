```python
df = pd.read_json("sample.json")
print(df.head())
```

```python
from pandas import json_normalize

df_orders = json_normalize(df.to_dict('records'), record_path=['orders'])
print(df_orders)
```

```python
order_totals = df_orders[['order_id', 'total']]
display(order_totals)
```

```python
from pandas import json_normalize

df_flat = json_normalize(df.to_dict('records'), record_path=['orders', 'items'],
meta=['customer_id', ['orders', 'order_id'], ['orders', 'total']])

df_flat.rename(columns={'orders.order_id': 'order_id', 'orders.total': 'total'}, inplace=True)
df_flat = df_flat[['customer_id','order_id','product','quantity','total']]
display(df_flat)
```

```python
df_flat.to_json("processed_orders.json", orient="records", indent=4)
```
